```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
```

```python
np.random.seed(42)  ## .Seed Retains the generated value

x = np.random.rand(50, 1) * 100

y = 3.5 * x + np.random.randn(50, 1) * 20
```

```python
x.shape
```

```python
## Creating and training Linear regression model

model = LinearRegression()

model.fit(x, y)
```

```python
## To predict value

y_pred = model.predict(x)
```

```python
## Visualizing the regression line
plt.figure(figsize=(8,6))
plt.scatter(x, y, color="blue", label="Data Points")
plt.plot(x, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression sample on Random dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Slope or Cosfficient 

print("Slope (Coefficient):", model.coef_[0] [0])
print("Intercept:", model.intercept_[0])
```

```python
## Slope or Cosfficient 

model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

# Slope (Coefficient)
slope = model.coef_[0]

print ("slope=", slope)
```
