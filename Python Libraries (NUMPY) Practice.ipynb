```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=60, scale=10, size=100)

plt.hist(data)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram of Normally Distributed Data by Paul')
plt.show()
```

```python
from scipy.stats import shapiro
```

```python
stat, p_value = shapiro(data)
print('shapiro_wilk test statistics:', stat)
print('p-value, p-value')

alpha = 0.05
if p_value > alpha:
    print('pass i.e fail to reject null hypothese')
else:
    print('fail')
```
