```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diabetes = pd.read_csv("../dataset/diabetes.csv")

print (diabetes)
```

```python
sns.set_style("whitegrid")
```

```python
## Scatter plot with sns 
sns.relplot(x = "BMI", y = "Age", data = diabetes)
```

```python
## Scatter plot with sns 
sns.relplot(x = "BMI", y = "BloodPressure", data = diabetes)

## for title add label
plt.xlabel("BMI")
plt.ylabel("Blood Pressure");  ### Semicolon to be added to remove numbers at the heading
```

```python
diabetes.loc[:, ["BMI", "BloodPressure", "Age"]].corr()

# This code is to find correlation
```

```python
sns.lmplot(x = "Pregnancies",y= "Age", data = diabetes);

# The blur line indicates the confidence interval
```

```python
# Pair Plot : Pairing variables against one another
```

```python
sns.pairplot (diabetes);
```

```python
## Join Plot: Scatter diagram with distribution of each variable
```

```python
sns.jointplot(x= "Glucose", y= "Insulin", data= diabetes);
```

```python
## Histogram
sns.histplot(x= diabetes["BMI"]);
```

```python
sns.histplot(x= diabetes["DiabetesPedigreeFunction"],kde=True);
```

```python
## Box Plot: To idenytify outliers in plot
```

```python
sns.boxplot(y="Pregnancies", data = diabetes);
## Outliers are above the data range
```

```python
## Count Plot 
```

```python
sns.countplot(x="Pregnancies", data = diabetes);  ## Count plot automatically counts in frequency
```

```python
sns.barplot(x="Outcome", y="Age", errorbar=None, data=diabetes, hue= "Outcome" );
```

```python
## Piechart 
diabetes["Pregnancies"].value_counts().plot(kind = "pie");
plt.xlabel("Pregnancies")
plt.ylabel("Outcome");
```
