```python
###  Class & Innit function 
class person():
    def __init__(self, name, age, gender,national, email):
        self.name = name
        self.age = age
        self.gender = gender
        self.national = national
        self.email = email 
    
    def describe_person(self):
        print (f"My name is {self.name}. I am a {self.gender}. I am {self.age}years old. My email is {self.email}")
    def nationality(self):
        print (f"my natioanlity is {self.national}")
    def department (self, dept_name, level):
        print(f"my department {dept_name}. I belong to {level}")
```

```python
result = person("Paul", 19, "Male", "Niger", "paulayodele04@gmail.com")
```

```python
result.describe_person()
```

```python
result.nationality()
```

```python
result.department('science',2)
```

```python
## Variable in a class
class residents:
    residence = 'Nigeria'

    def __init__(self, name, nationality):
        self.name = name
        self.nationality = nationality


    def describe_me(self):
        print (f"My name is {self.name}. I am a {self.nationality}  I live in {residents.residence}.")

```

```python
r = residents('Paul', 'Nigerian')
r.describe_me()
```

```python
## Bank account deposit & Withdrawal
class Bank_Account:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit (self, amount):
        if amount < 0:
            print ('Amount unacceptable')
        else:
            self.balance += amount
    def withdraw(self, amount):
        self.balance -+ amount
    
    def check_balance(self):
        print (f"my account balance is {self.balance}")
    
```

```python
bank = Bank_Account('Alice', 1000)
```

```python
bank.deposit(500)
```

```python
bank.withdraw(200)
```

```python
bank.check_balance()
```
